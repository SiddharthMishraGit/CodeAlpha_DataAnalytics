```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set default Seaborn style
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python
df.shape  # rows & columns
df.info()  # data types + nulls
df.describe()  # summary stats
df.isnull().sum()  # missing values
```

```python
# Survival count
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
```

```python
# Gender
sns.countplot(x='Sex', data=df)
plt.title("Passenger Gender Count")
```

```python
sns.countplot(x='Pclass', data=df)
plt.title("Passenger Class Count")
```

```python
# Gender vs Survival
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Gender")
```

```python
# Class vs Survival
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival by Class")
```

```python
# Age Distribution
sns.histplot(df['Age'].dropna(), kde=True)
plt.title("Age Distribution")
```

```python
# Age vs Survival (Boxplot)
sns.boxplot(x='Survived', y='Age', data=df)
plt.title("Age Distribution by Survival")
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

### 🔍 Insights & Observations

- Female passengers had a much higher survival rate than males.
- 1st class passengers were more likely to survive compared to 2nd and 3rd class.
- Younger passengers (especially children) had higher chances of survival.
- Fare and class had a strong positive correlation with survival.
- No missing data was found in this version of the dataset, so no imputation was needed.

This EDA provided valuable insights into how socio-demographic factors influenced survival during the Titanic disaster.
